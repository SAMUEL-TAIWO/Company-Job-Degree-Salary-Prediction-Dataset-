```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import warnings
warnings.filterwarnings("ignore")
```

```python
salaries=pd.read_csv("salaries.csv")
salaries
```

```python
input=salaries.drop("salary_more_then_100k",axis = 1)
input
```

```python
target = salaries[["salary_more_then_100k"]]
target
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
input
```

```python
#using fit transform
input["company_x"]=le_company.fit_transform(input["company"])
input["job_x"]=le_job.fit_transform(input["job"])
input["degree_x"]=le_degree.fit_transform(input["degree"])
```

```python
input
```

```python
#drop the initial columns(company	job	degree	)
final_dummy =input.drop(["company","job","degree"], axis = 1)
final_dummy
```

```python
target
```

```python
model = tree.DecisionTreeClassifier()
model
```

```python
model.fit(final_dummy,target)
```

```python
model.score(final_dummy,target)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.datasets import load_digits
sns.set 
digits = load_digits()
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
dir(digits)
```

```python
df=pd.DataFrame(digits.data)
df.head()
```

```python
df["targets"]=digits.target
```

```python
df[0:12]
```

```python
x=df.drop("targets",axis = 1)
y=df.targets
```

```python
X_train, X_test, y_train, y_test=train_test_split(x,y, test_size = 0.2)
```

```python
model = RandomForestClassifier( n_estimators = 20)
model.fit(X_train,y_train)
```

```python
y_test
```

```python
X_test
```

```python
model.score(X_test,y_test)
```
